```python
target_col = 'close'

def line_plot(line1, line2, label1=None, label2=None, title='', lw=2):
   fig, ax = plt.subplots(1, figsize=(16, 9))
   ax.plot(line1, label=label1, linewidth=lw)
   ax.plot(line2, label=label2, linewidth=lw)
   ax.set_ylabel('price [USD]', fontsize=14)
   ax.set_title(title, fontsize=18)
   ax.legend(loc='best', fontsize=18)

line_plot(hist_BTC[target_col], hist_LTC[target_col], 'BTC', 'LTC', title='BTC vs LTC')
```

### Our Strategy:
__Convert the time series into a more stationary time series so the model can learn better from it__

### How to convert?
 - __Differencing:__ one common strategy is to use price of x(t) - x(t-1) to convert a nonstationary into a stationary time series
 - __Differencing with other assets:__another strategy is to calculate the price difference of 2 different coins to get stationary time series
 - __Decomposing:__ decompose time series into trend, seasonal, and residual. we use residual to predict


```python
s_test = hist_LTC[target_col]
log_return = np.log(s_test/s_test.shift())
log_return = log_return.dropna()
plt.plot(log_return)
plt.title('log return of LiteCoin (x(t) - x(t-1)')
plt.show()
```

```python
s2 = hist_LTC[target_col]/hist_BTC[target_col]
#s2 = hist_BTC[target_col]/hist_LTC[target_col]
plt.plot(s2)
plt.title('LTC over BTC')
plt.show()
```

```python
s3 = np.log(hist_LTC[target_col]) # we use log price

from statsmodels.tsa.seasonal import seasonal_decompose
decomposition = seasonal_decompose(s3)
trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

plt.subplot(411)
plt.plot(s3, label = 'Original')
plt.legend(loc = 'best')
plt.subplot(412)
plt.plot(trend, label = 'Trend')
plt.legend(loc = 'best')
plt.subplot(413)
plt.plot(seasonal, label = 'Seasonality')
plt.legend(loc = 'best')
plt.subplot(414)
plt.plot(residual, label = 'Residual')
plt.legend(loc = 'best')
```

__even the residual looks stationary, but we think it is hard to handle the trend data, so we decide to test the stationarity for only first 2 methods__
### Check for stationarity for above first 2 strategies

 - It is easier when the time series is stationary: mean-reversion
 - __Augmented Dickey - Fuller test__ is commonly used to check stationarity in a time series.
 - Stationarity is an important property to make sure what happened in the past is more likely to repeat in the future

```python
from statsmodels.tsa.stattools import adfuller

def test_stationarity(timeseries):
    
    #Determining rolling statistics
    rolmean = timeseries.rolling(window = 12).mean()
    rolstd = timeseries.rolling(window = 12).std()
    
    #plotting rolling statistics
    orig = plt.plot(timeseries, color = 'blue', label = 'Original')
    mean = plt.plot(rolmean, color = 'red', label = 'Rolling Mean')
    st = plt.plot(rolstd, color = 'black', label = 'Rolling Std')
    plt.legend(loc = 'best')
    plt.title('Rolling  Mean & Standard Deviation')
    plt.show()
    
    #Dickey_Fuller test:
    print('Results of Dickey-Fuller Test:')
    dftest = adfuller(timeseries)
    dfoutput = pd.Series(dftest[0:4], index = ['Test Statistic', 'p-value', '#Lags Used', 'Number of Observations Used'])
    for key, value in dftest[4].items():
        dfoutput['Critical Value (%s)' %key] = value
    print(dfoutput)
```

```python
test_stationarity(log_return)
```

### Takeaway from above:
 - the p-value is way below 0.05, so we can say it is a stationary time series

```python
test_stationarity(s2)
```

### Takeaway:
 - the p-value is 0.14, so we are not able to reject the H0, and thus the time series is non-stationary
 - we might use method one (differencing) to feed into deep learning model to predict log return instead of actual coin price

### What can we do with s2?
 - __Arbitrage:__ pairs trading
 - As we know, cryptocurrency has gone through a major increase and a major decrease in the past few years. Is there any strategy to make money no matter if coin price goes up or goes down? 
 - Since coins are similar assets, so we assume they have similar behaviors and trends, and their prices are exposed to similar factors. We think the price of 2 coins are likely to be correlated and cointegrated. These properties are perfect for pairs trading.
 - Pairs trading is a market neutral strategy where there is little exposure to the direction of the overall market.

```python
# first we standardize s2
def standardize(series):
    return (series - series.mean()) /np.std(series)


```

### Take a look of the day-level data of LTC over BTC
 - after standadization, for most of the time, the time series fluctuates between upperbound of 1, and lowerbound of -1
 - whenever it goes higher than 1, LTC is likely to be overpriced and BTC is likely to be underpriced. The relative price will soon come back to be within [-1,1]. Short LTC and long BTC
 - whenever it goes below than -1, LTC is likely to be underpriced and BTC is likely to be overpriced. The relative price will soon come back to be within [-1,1]. Long LTC and short BTC
